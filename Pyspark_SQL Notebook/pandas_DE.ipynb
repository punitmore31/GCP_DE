```python
import json

# The sample JSON data stored as a multi-line string
json_string = """
{
  "name": "Alex Doe",
  "age": 22,
  "isStudent": true,
  "courses": [
    "History",
    "Computer Science",
    "Art"
  ],
  "address": {
    "city": "Pune",
    "zipcode": "411001"
  }
}
"""

# 1. Parse the JSON string into a Python dictionary
# The `json.loads()` function (load string) does this conversion.
user_data = json.loads(json_string)
print(f"Type of parsed data: {type(user_data)}")
# Output: Type of parsed data: <class 'dict'>
print("-" * 20)


# 2. Access data from the dictionary
# You can access values using their keys.
print('Users Name:   ',   user_data['name'])
print(f"User's City: {user_data['address']['city']}") # Accessing a nested value
print(f"First Course: {user_data['courses'][0]}")   # Accessing an item in a list
print("-" * 20)


# 3. Modify the data
# You can change values, add new key-value pairs, or append to lists.
user_data['age'] = 23  # Update a value
user_data['email'] = "alex.doe@example.com"  # Add a new key-value pair
user_data['courses'].append("Mathematics")  # Add an item to the list

print("Modified user data:")
print(user_data)
# Output:
# Modified user data:
# {'name': 'Alex Doe', 'age': 23, 'isStudent': True, 'courses': ['History', 'Computer Science', 'Art', 'Mathematics'], 'address': {'city': 'Pune', 'zipcode': '411001'}, 'email': 'alex.doe@example.com'}
print("-" * 20)


# 4. Convert the Python dictionary back to a JSON string
# The `json.dumps()` function (dump string) does this.
# `indent=4` makes the output nicely formatted and readable ("pretty-printed").
new_json_string = json.dumps(user_data, indent=4)

print("Newly generated JSON string:")
print(new_json_string)
```

```python
# Calculate a 3-day rolling mean of Sales
df 
df['3D_Moving_Avg'] = df.groupby('order_id')['order_id'].rolling(window=3).sum().reset_index(level=0, drop=True)

print("\nDataFrame with 3-Day Moving Average:")
print(df)
```

```python
# Calculate the expanding (cumulative) sum of Sales for each product
df['Cumulative_Sales'] = df.groupby('Product')['Sales'].expanding().sum().reset_index(level=0, drop=True)

print("\nDataFrame with Cumulative Sales:")
print(df.drop(columns='3D_Moving_Avg')) # Dropping previous column for clarity

# Rank sales within each product group, with the highest sale being rank 1
df['Sales_Rank'] = df.groupby('Product')['Sales'].rank(method='dense', ascending=False)

print("\nDataFrame with Sales Rank per Product:")
print(df.drop(columns=['3D_Moving_Avg', 'Cumulative_Sales']))

# First, get the previous day's sales
df['Previous_Day_Sales'] = df.groupby('Product')['Sales'].shift(1)

# Then, calculate the difference
df['Sales_Change'] = df['Sales'] - df['Previous_Day_Sales']

print("\nDataFrame with Day-over-Day Sales Change:")
print(df.drop(columns=['3D_Moving_Avg', 'Cumulative_Sales', 'Sales_Rank']))
```
