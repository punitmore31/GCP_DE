```python
def sieve_of_eratosthenes(N):
    primes = [True] * N  # Assume all numbers are prime
    primes[0] = primes[1] = False  # 0 and 1 are not prime
    
    for i in range(2, int(N ** 0.5) + 1):
        if primes[i]:
            for j in range(i * i, N, i):  # Mark multiples of i as non-prime
                primes[j] = False
    
    return [i for i in range(N) if primes[i]]

N = int(input("Enter a number N: "))
print(f"Prime numbers less than {N}: {sieve_of_eratosthenes(N)}")
```

```python
#print all the prime number from 20 to 50
def is_prime(n):
    if n == 1:
        print(f'{n} is not prime number')
        return False
    else:
        if n>1:
            for i in range(2,n):
                if n % i == 0:
                    return False
            return True


prime_list = []
non_prime_list = []
for num in range(20, 30):
    if is_prime(num):
        prime_list.append(num)
    else :
        non_prime_list.append(num)

print(f"prime number list Bet 20 & 40 : {prime_list}")
print(f"Non prime number list Bet 20 & 40 : {non_prime_list}")
```

```python
# 1. Generate an Infinite Fibonacci Series Using a Generator


def infinite_fibonacci():
    # Initialize the first two numbers
    a, b = 0, 1
    while True:
        # 'yield' makes this a generator
        yield a
        # Update the numbers for the next iteration
        a, b = b, a + b


# Example usage: Get the first 10 numbers
fib_gen = infinite_fibonacci()
print("First 10 Fibonacci numbers:")
for _ in range(10):
    print(next(fib_gen), end=" ")
print("\n---")
```

```python
# 2. Sort a List Without Using the Sort Method


def bubble_sort(data_list):
    n = len(data_list)
    # Traverse through all elements in the list
    for i in range(n):
        # Last i elements are already in place, so we don't need to check them
        for j in range(0, n - i - 1):
            # Swap if the element found is greater than the next element
            if data_list[j] > data_list[j + 1]:
                data_list[j], data_list[j + 1] = data_list[j + 1], data_list[j]
    return data_list


my_list = [64, 34, 25, 12, 22, 11, 90]
sorted_list = bubble_sort(
    my_list.copy()
)  # Use a copy to keep the original list unchanged
print("Original list:", my_list)
print("Sorted list (Bubble Sort):", sorted_list)
print("---")
```

```python
# 3. Check if a String is a Palindrome


def is_palindrome(s):
    # Remove spaces and convert to lowercase for a case-insensitive, space-ignoring check
    s = s.replace(" ", "").lower()
    # Check if the string is equal to its reverse
    # s[::-1] creates a reversed copy of the string
    return s == s[::-1]


print("'madam' is palindrome:", is_palindrome("madam"))
print(
    "'A man a plan a canal Panama' is palindrome:",
    is_palindrome("A man a plan a canal Panama"),
)
print("'hello' is palindrome:", is_palindrome("hello"))
print("---")
```

```python
# 4. Find the Factorial of a Number


def find_factorial(n):
    # Factorial is only defined for non-negative integers
    if n < 0:
        return "Factorial is not defined for negative numbers"
    elif n == 0:
        return 1
    else:
        factorial = 1
        # Loop from 1 up to n (inclusive)
        for i in range(1, n + 1):
            factorial = factorial * i
        return factorial


number = 5
print(f"The factorial of {number} is:", find_factorial(number))
number_zero = 0
print(f"The factorial of {number_zero} is:", find_factorial(number_zero))
print("---")
```

```python
# 5. Count Word Frequency in a Sentence


def count_word_frequency(sentence):
    # Convert to lowercase and split the sentence into a list of words
    words = sentence.lower().split()
    print(f"Words is : {words}")
    word_counts = {}

    for word in words:
        # Simple cleanup: remove punctuation like periods or commas
        word = word.strip(".,!?\"'")
        print(f"word = {word}")
        if word:  # Ensure the cleaned word is not empty
            # If the word is already a key, increment its count
            if word in word_counts:
                word_counts[word] += 1
            # Otherwise, add the word to the dictionary with a count of 1
            else:
                word_counts[word] = 1

    return word_counts


my_sentence = "Python is great. Python  is fun. and Python is. useful."
frequency_map = count_word_frequency(my_sentence)
print("Word frequency count:", frequency_map)
print("---")
```

```python
# 6. Find the Smallest and Largest Number in an Array (List)


def find_min_max(arr):
    if not arr:
        return None, None  # Return None if the list is empty

    # Initialize min_num and max_num with the first element
    min_num = arr[0]
    max_num = arr[0]

    # Iterate through the rest of the list starting from the second element
    for number in arr[1:]:
        if number < min_num:
            min_num = number  # Found a new minimum
        if number > max_num:
            max_num = number  # Found a new maximum

    return min_num, max_num


number_array = [45, 12, 89, 2, 77, 34, 1]
smallest, largest = find_min_max(number_array)

print("Array:", number_array)
print(f"Smallest number: {smallest}")
print(f"Largest number: {largest}")
```

```python
# 7. Find the Top N Records (without using Pandas)


def find_top_n_records(data_list, n, key_index, reverse=True):
    """
    Finds the top N records based on a specific key in a list of lists/tuples.
    """
    if not data_list:
        return []

    # Sort the list based on the element at key_index.
    # The 'reverse=True' sorts it in descending order (for 'top N').
    sorted_data = sorted(data_list, key=lambda x: x[key_index], reverse=reverse)

    # Return the first N elements
    return sorted_data[:n]


# Sample data: (User ID, Sales Amount, Region)
sales_data = [
    (101, 5000, "North"),
    (102, 12000, "South"),
    (103, 3000, "North"),
    (104, 18000, "West"),
    (105, 7500, "East"),
]

# Find the top 2 records based on Sales Amount (index 1)
top_sales = find_top_n_records(sales_data, 2, key_index=1)
print("Top 2 Sales Records (by Amount):")
for record in top_sales:
    print(record)
print("---")

```

```python
# 8. Implement Run-Length Encoding (RLE)


def run_length_encode(data_string):
    """
    Compresses a string using Run-Length Encoding.
    Example: 'AAABBC' -> ['A', 3, 'B', 2, 'C', 1]
    """
    if not data_string:
        return []

    encoded = []
    i = 0
    n = len(data_string)

    while i < n:
        current_char = data_string[i]
        count = 1
        j = i + 1

        # Count consecutive identical characters
        while j < n and data_string[j] == current_char:
            count += 1
            j += 1

        # Append the character and its count to the result
        encoded.append((current_char, count))

        # Move the main pointer to the start of the next sequence
        i = j

    return encoded


input_data = "AAAAABBBCCDAAA"
encoded_result = run_length_encode(input_data)
print(f"Original: '{input_data}'")
print("RLE Output:", encoded_result)
print("---")
```

```python
# 9 . Merge Two Sorted Lists


def merge_sorted_lists(list1, list2):
    """
    Merges two sorted lists into a single sorted list efficiently.
    This runs in O(n+m) time, where n and m are the lengths of the lists.
    """
    merged = []
    i = j = 0  # Pointers for list1 and list2

    # While both lists have elements remaining
    while i < len(list1) and j < len(list2):
        if list1[i] <= list2[j]:
            merged.append(list1[i])
            i += 1
        else:
            merged.append(list2[j])
            j += 1

    # Append any remaining elements from list1 (if any)
    merged.extend(list1[i:])
    # Append any remaining elements from list2 (if any)
    merged.extend(list2[j:])

    return merged


list_a = [1, 5, 8, 10]
list_b = [2, 4, 9, 11, 15]
merged_list = merge_sorted_lists(list_a, list_b)
print("List A:", list_a)
print("List B:", list_b)
print("Merged List:", merged_list)
print("---")

```

```python
# 10. Group Data by Key (Equivalent to SQL GROUP BY)

from collections import defaultdict


def group_data_by_key(data, key_index):
    """
    Groups records in a list of lists/tuples based on a specific key index.
    """
    # Use defaultdict to automatically create a list for a new key
    grouped_data = defaultdict(list)

    print(f"grouped data : {grouped_data}")

    for record in data:
        key = record[key_index]
        grouped_data[key].append(record)

    # Convert back to a regular dict for final output
    return dict(grouped_data)



# Sample data: (ID, Name, Department, Salary)
employee_data = [
    (1, "Alice", "Engineering", 70000),
    (2, "Bob", "HR", 60000),
    (3, "Charlie", "Engineering", 85000),
    (4, "Diana", "HR", 62000),
    (5, "Eve", "Marketing", 55000),
]

# Group data by Department (index 2)
grouped_by_dept = group_data_by_key(employee_data, key_index=2)
print("Records Grouped by Department:")
for dept, records in grouped_by_dept.items():
    print(f"   {dept}: {len(records)} employees")
    # print(f"    {records}")
print("---")

```

```python
# 11. Find Missing Number in a Sequence


def find_missing_number(arr, n):
    """
    Finds the missing number in an array of size (n-1) that should contain
    numbers from 1 to n (inclusive).
    Uses the sum of an arithmetic series formula for O(N) efficiency.
    """
    # 1. Calculate the expected sum of all numbers from 1 to n
    # The formula for the sum of the first n integers is: Sum = n * (n + 1) / 2
    expected_sum = n * (n + 1) // 2

    # 2. Calculate the actual sum of the numbers present in the array
    actual_sum = sum(arr)

    # 3. The missing number is the difference
    missing_number = expected_sum - actual_sum

    return missing_number


# Array has numbers from 1 to 9, so n = 10 (the complete range)
data_array = [1, 2, 4, 5, 6, 7, 8, 9, 10]
n_range = 10
missing = find_missing_number(data_array, n_range)
print(f"The sequence should contain 1 to {n_range}.")
print(f"Actual array: {data_array}")
print("Missing number:", missing)
print("---")
```

```python
# 13. Find Duplicates in a List


def find_duplicates(data_list):
    """
    Finds all elements that appear more than once in the list.
    """
    seen = set()
    duplicates = set()

    for item in data_list:
        if item in seen:
            duplicates.add(item)
        else:
            seen.add(item)

    return list(duplicates), list(seen)
    


data_with_dupes = [10, 20, 30, 10, 40, 50, 20, 60, 30, 30]
duplicate_items = find_duplicates(data_with_dupes)
dupes, unique_items = find_duplicates(data_with_dupes)

print("Original List:", data_with_dupes)
print("Duplicate items found:", dupes)
print("Duplicate items found:", sorted(unique_items))
```
