```python
%sql
select id,
concat(
  split(contact_number,' ')[0], 
  repeat("*", length(split(contact_number,' ')[1]) - 4),
  Right(contact_number,4)
  ) as Masked_phone_number,
  mail_id,
  split(split(mail_id, '@')[0],'\\.')[0] as First_name, --In Spark SQL (and many other SQL dialects), the second argument of the split() function is treated as a Regular Expression (Regex), not just a plain text character. And dot have special mening in regex
  split(split(mail_id, '@')[0],'\\.')[1] as Last_name,
  status
 from employee;

```

```python
from pyspark.sql import SparkSession
from pyspark.sql.functions import col, split, concat, expr

# Initialize Spark session (skip this line if running in Databricks where 'spark' is already defined)
spark = SparkSession.builder.appName("EmployeeMasking").getOrCreate()

# ---------------------------------------------------------
# 1. Create the DataFrame (Replaces CREATE TABLE & INSERT)
# ---------------------------------------------------------
data = [
    (1, '+91 9876543210', 'john.doe@gmail.com', 'Active'),
    (2, '+1 9123456780',  'jane.smith@gmail.com', 'Inactive'),
    (3, '+44 9988776655', 'mike.brown@gmail.com', 'Active'),
    (4, '+91 9012345678', 'emily.white@gmail.com', 'Inactive'),
    (5, '+61 9090909090', 'robert.black@gmail.com', 'Active'),
    (6, '+1 8887776665',  'sara.green@gmail.com', 'Active')
]

columns = ["id", "contact_number", "mail_id", "status"]

employee_df = spark.createDataFrame(data, columns)

# ---------------------------------------------------------
# 2. Apply Transformations (Replaces SELECT query)
# ---------------------------------------------------------
transformed_df = employee_df.select(
    col("id"),
    
    # Masking the phone number
    # We use expr() for repeat and right because PySpark's native repeat() 
    # expects a hardcoded integer, and right() is easiest as a SQL expression.
    concat(
        split(col("contact_number"), ' ').getItem(0), 
        expr("repeat('*', length(split(contact_number, ' ')[1]) - 4)"),
        expr("right(contact_number, 4)")
    ).alias("Masked_phone_number"),
    
    col("mail_id"),
    
    # Extracting First Name (escaping the dot)
    split(split(col("mail_id"), '@').getItem(0), '\\.').getItem(0).alias("First_name"),
    
    # Extracting Last Name (escaping the dot)
    split(split(col("mail_id"), '@').getItem(0), '\\.').getItem(1).alias("Last_name"),
    
    col("status")
)

# Display the results
display(transformed_df) # Use transformed_df.show() if running in a standard terminal
```

######### 👉 3. Find employees who logged in for *3 consecutive days* using:  
#########✅ Databricks SQL    
########✅ PySpark (Spark SQL / DataFrame API)  

```python
data = [
    (1, '1/20/2025'),
    (1, '1/21/2025'),
    (1, '1/22/2025'),
    (1, '1/24/2025'),
    (2, '1/15/2025'),
    (2, '1/16/2025'),
    (2, '1/18/2025'),
    (3, '1/15/2025'),
    (3, '1/16/2025'),
    (3, '1/17/2025'),
    (3, '1/18/2025')
]

columns = ["emp_id", "login_date"]

```

#### Solution 1: Without using the LEAD `function`

```python
from pyspark.sql import SparkSession
from pyspark.sql.functions import *
from pyspark.sql.window import Window
from pyspark.sql.types import StructType, StructField, StringType, IntegerType, DateType

df = spark.createDataFrame(data, columns)
# display(df)


cleaned_df = df.dropDuplicates(["emp_id","login_date"])

grouped_df = cleaned_df.withColumn('rn', row_number().over(Window.partitionBy("emp_id").orderBy("login_date"))).withColumn("login_date", to_date(col("login_date"), "M/d/yyyy").cast(DateType()))

grouped_df1 = grouped_df.withColumn("group_id", expr("date_sub(login_date,rn)")).groupBy("emp_id","group_id").agg(count("*").alias('count')).filter("count>=3")

display(grouped_df1)
```

#### Solution 2: With the LEAD `function`

```python
from pyspark.sql import SparkSession
from pyspark.sql.functions import col, to_date, lead, datediff
from pyspark.sql.window import Window

# 1. Initialize session and create DataFrame
spark = SparkSession.builder.appName("ConsecutiveLogins").getOrCreate()

data = [
    (1, '1/20/2025'), (1, '1/21/2025'), (1, '1/22/2025'), (1, '1/24/2025'),
    (2, '1/15/2025'), (2, '1/16/2025'), (2, '1/18/2025'),
    (3, '1/15/2025'), (3, '1/16/2025'), (3, '1/17/2025'), (3, '1/18/2025')
]
columns = ["emp_id", "login_date"]
df = spark.createDataFrame(data, columns)

# 2. Convert string to actual Date type and drop duplicates 
# (in case an employee logs in multiple times on the same day)
df_clean = (
    df.withColumn("login_date", to_date(col("login_date"), "M/d/yyyy"))
    .dropDuplicates(["emp_id", "login_date"])
)

# 3. Define Window specification
window_spec = Window.partitionBy("emp_id").orderBy("login_date")

# 4. Apply LEAD function and filter for 3 consecutive days
consecutive_emps_df = (
    df_clean
    # Look 2 rows ahead to find the login date 2 sequences away
    .withColumn("date_plus_2", lead("login_date", 2).over(window_spec))
    # If the date 2 rows ahead is exactly 2 days after the current row, it's a 3-day streak
    .filter(datediff(col("date_plus_2"), col("login_date")) == 2)
    # Select only the distinct employee IDs
    .select("emp_id")
    .distinct().limit(1)
)

display(consecutive_emps_df)

df.write.mode("overwrite").saveAsTable("employee_logins")
```

```python
%sql
WITH CleanedDates AS (
    -- Convert string to date and ensure 1 record per employee per day
    SELECT DISTINCT 
        emp_id, 
        to_date(login_date, 'M/d/yyyy') AS login_date
    FROM employee_logins
),
LeadDates AS (
    -- Use LEAD to peek 2 rows ahead within each employee's partition
    SELECT 
        emp_id,
        login_date,
        LEAD(login_date, 2) OVER (PARTITION BY emp_id ORDER BY login_date) AS date_plus_2
    FROM CleanedDates
)
-- Filter where the gap between the current date and the date 2 rows down is exactly 2 days
SELECT DISTINCT emp_id
FROM LeadDates
WHERE datediff(date_plus_2, login_date) = 2;
```

```python
data = [
    (1, "+91 9876543210", "john.doe@gmail.com", "Active"),
    (2, "+1 9123456780", "jane.smith@gmail.com", "Inactive"),
    (3, "+44 9988776655", "mike.brown@gmail.com", "Active"),
    (4, "+91 9012345678", "emily.white@gmail.com", "Inactive"),
    (5, "+61 9090909090", "robert.black@gmail.com", "Active")
]
columns = ["id", "contact_number", "mail_id", "status"]

employee_df = spark.createDataFrame(data, columns)

# 📌 Country Dictionary (Mapping Source)
country_dict = {
    "+91": "India",
    "+1": "United States",
    "+44": "United Kingdom",
    "+61": "Australia"
}
```

```python
# More efficient for "Top N" problems
pickup_Final_result_df = (pickup_df
    .groupBy("pickup_location")
    .count() # Shortcut for agg(count("*"))
    .orderBy(col("count").desc())
    .limit(3)
    .select("pickup_location"))

display(pickup_Final_result_df)
```

#### Problems 14 : Calculate the start and end date of show status

```python
from pyspark.sql import SparkSession
from pyspark.sql.functions import *
from pyspark.sql.types import StructType, StructField, StringType
from pyspark.sql.window import Window

data = [
    ("01-06-2020","Booked"),
    ("02-06-2020","Booked"),
    ("03-06-2020","Booked"),
    ("04-06-2020","Available"),
    ("05-06-2020","Available"),
    ("06-06-2020","Available"),
    ("07-06-2020","Booked")
]

schema = StructType([
    StructField("show_date", StringType(), True),
    StructField("show_status",StringType(), True)
])

spark = SparkSession.builder.appName("Solutions").getOrCreate()
df = spark.createDataFrame(data, schema)
df.display()

df = (
    df.withColumn("show_date", to_date(col("show_date"),"dd-MM-yyyy"))
)

df = df.withColumn("show_change", lag(col("show_status")).over(Window.orderBy(col("show_date"))))
display(df)

df = df.withColumn("show_change", when(col("show_status") != lag(col("show_status")).over(Window.orderBy(col("show_date"))),1).otherwise(0))

display(df)

#Create a column show_group based on cumulative sum of show_change
df1 = df.withColumn("show_group",sum("show_change").over(Window.orderBy(col("show_date"))))

display(df1)

result_df = df1.groupBy("show_group","show_status").agg(min("show_date").alias("start_date"), max("show_date").alias("end_date")).orderBy("start_date")

display(result_df)

```

#####Problem 18 : Get Customer Journey Location

```python
from pyspark.sql.functions import *
from pyspark.sql.types import *

data = [
    ('A001', 'Tokyo', 'Seoul'),
    ('A001', 'Singapore', 'Tokyo'),
    ('A001', 'Seoul', 'Bangkok'),
    ('A001', 'Bangkok', 'Manila'),
    ('B002', 'Paris', 'Rome'),
    ('B002', 'Berlin', 'Paris'),
    ('B002', 'Rome', 'Madrid'),
    ('C003', 'Cairo', 'Dubai'),
    ('C003', 'Istanbul', 'Cairo'),
    ('C003', 'Dubai', 'Riyadh'),
    ('C003', 'Riyadh', 'Doha'),
    ('D004', 'Toronto', 'Montreal'),
    ('D004', 'Vancouver', 'Toronto'),
    ('E005', 'Sydney', 'Melbourne')
]

# Define schema
schema = StructType([
    StructField("customer", StringType(), True),
    StructField("start_location", StringType(), True),
    StructField("end_location", StringType(), True)
])

# Create DataFrame
df = spark.createDataFrame(data=data, schema=schema)

print("Original DataFrame:")
df.display()


df_start = df.select("customer",col("start_location").alias("location"))
df_end = df.select("customer",col("end_location").alias("location"))

# display(df_start)
# display(df_end)


ultimate_start = df_start.join(df_end, on = ["location", "customer"], how="left_anti")
ultimate_end = df_end.join(df_start, on=["location","customer"],how="left_anti")

final_df = ultimate_start.join(ultimate_end, on = ["customer"], how="inner")

display(final_df)


```

######problem 19 : Swap Consecutive Names in PySpark

```python
from pyspark.sql.functions import *
from pyspark.sql.window import Window
data = [
    (1,"Alice", 1200, "Q1"),
    (2,"Bob", 900, "Q1"),
    (3,"Charlie", 1500, "Q2"),
    (4,"David", 1700, "Q2"),
    (5,"Eva", 1100, "Q3"),
    (6,"Frank", 220, "Q3"),
    (7,"Grace", 1300, "Q4")
]


df = spark.createDataFrame(data, ["id", "name", "sales", "quarter"])
df.display()

df = df.withColumn("previous",lag("name").over(Window.orderBy(asc("id"))))
df = df.withColumn("next",lead("name").over(Window.orderBy(asc("id"))))

df_interchange = df.withColumn(
    "interchange", 
    when(col("id") % 2 != 0, coalesce(col("next"), col("name"))).when(col("id") % 2 == 0, col("previous"))
)

# df_result = df_interchange.selectExpr("id","name","interchange")
df_result = df_interchange.selectExpr("id","name","interchange")

display(df_result)
```

######Problems 21(Deloitte) : Write Pysparks query and sql query to find the products whose total sales revenue increased every year. Include product_id, name , category in your result dataset.

```python
from pyspark.sql.functions import *
from pyspark.sql.window import *
product_data = [
    (1, 'Laptops', 'Electronics'),
    (2, 'Jeans', 'Clothing'),
    (3, 'Chairs', 'Home Appliances')
]

product_schema = ['product_id', 'product_name', 'category']

product_df = spark.createDataFrame(data=product_data, schema=product_schema)
print("Product DataFrame:")
# product_df.display()

sales_data = [
    (1, 2019, 1000.00),
    (1, 2020, 1500.00),
    (1, 2021, 1200.00),
    (2, 2019, 500.00),
    (2, 2020, 700.00),
    (2, 2021, 900.00),
    (3, 2019, 400.00),
    (3, 2020, 450.00),
    (3, 2021, 300.00)
]

sales_schema = ['product_id', 'year', 'total_sales_revenue']

sales_df = spark.createDataFrame(data=sales_data, schema=sales_schema)
print("Sales DataFrame:")
# sales_df.display() 


sales_df = sales_df.withColumn("prev_revenue",lag(col("total_sales_revenue")).over(Window.partitionBy(col("product_id")).orderBy(col("year")))).withColumn("min_revenue", col("total_sales_revenue")-col("prev_revenue")).groupBy(col("product_id")).agg(min(col("min_revenue")).alias("min_valuw")).filter(col("min_valuw")>0)

result_df = sales_df.join(product_df, ["product_id"],"inner").select("product_id","product_name","category")

display(result_df)
```
