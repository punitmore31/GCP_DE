```python
from pyspark.sql import SparkSession

# Initialize Spark Session (if not already done)
spark = SparkSession.builder.appName("CustomDelimiterLoad").getOrCreate()

file_path = "gs://my-bucket/data/input.csv"
custom_delimiter = "~|"

df = spark.read.csv(
    file_path,
    sep=custom_delimiter,  # <--- Crucial step: Specify the custom delimiter
    header=True,           # Assumes the first row is the header
    inferSchema=True,      # Spark attempts to guess the data types
    mode="FAILFAST"        # Ensures the job fails immediately on corrupt records
)

df.printSchema()
df.show(5)
```
