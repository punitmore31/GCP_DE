```python
# 4. re.finditer() - Find all matches as an iterator
# This is the memory-efficient version of findall(). Instead of a list, it gives you an iterator of match objects (like the one you got from re.search()).

# Use: Best for large files where you might find thousands of matches and don't want to load them all into memory at once.
import re 

content = """

/* HI THIS IS COMMENT PART*/

THIS IS NON COMMENT PART

"""

# Your regex pattern to find the comment
pattern = r'/\*.*?\*/'
flags = re.DOTALL

matches_iterator = re.finditer(pattern, content, flags=flags)

print("--- Using re.finditer() ---")
for match in matches_iterator:
    print(f"Found match: {match.group(0)} at index {match.start()}")
```
